```python
import requests

# Fetch data from Starbucks API
url = "https://www.starbucks.com/bff/ordering/menu"
response = requests.get(url)

# Print the full JSON response structure
print(response.json())
```

```python
import json

# Convert response to JSON
menu_data = response.json()

# Pretty print the structure
print(json.dumps(menu_data, indent=2))
```

```python
coffee_prices = []
for menu in menu_data['menus']:
    for category in menu['children']:  # Accessing categories
        for subcategory in category['children']:  # Accessing subcategories
            for product in subcategory['products']:  # Extracting products
                coffee_prices.append({
                    "Name": product["name"],
                    "Image": product["assets"]["masterImage"]["uri"],  # If you want images
                    "Sizes": [size["sizeCode"] for size in product["sizes"]],
                    "URL": f"https://www.starbucks.com{product['uri']}"
                })

import pandas as pd

df_coffee = pd.DataFrame(coffee_prices)
df_coffee.head()  # Display the first few rows
```

```python
import matplotlib.pyplot as plt

# Grouping by 'Coffee Type' and calculating the mean price
avg_price = df.groupby('Coffee Type')['Latte Price'].mean()

# Plotting the bar chart
plt.figure(figsize=(10, 5))
avg_price.plot(kind='bar', color='skyblue')
plt.xlabel('Coffee Type')
plt.ylabel('Average Price')
plt.title('Average Price of Coffee Types')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Rating'], df['Latte Price'], color='blue', alpha=0.5)
plt.xlabel('Rating')
plt.ylabel('Latte Price')
plt.title('Coffee Price vs. Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Grouping by 'City' and calculating the mean price
avg_price_city = df.groupby('City')['Latte Price'].mean()

# Plotting the bar chart
avg_price_city.sort_values().plot(kind='bar', color='orange')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Coffee Price by City')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))

# Pivot the data: Cities as rows, Coffee Types as columns, and Latte Price as values
heatmap_data = df.pivot_table(index='City', values='Latte Price', aggfunc='mean')

# Plot the heatmap
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".2f")

plt.title("Coffee Prices Across Cities")
plt.xlabel("Coffee Type")
plt.ylabel("City")
plt.show()
```
