```python
#1.Take tha data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv')
df
```

```python
#2.step no 2 not required
```

```python
#4.Divide the data into input and output
```

```python
#input - sepal length,sepal width, petal length,petal width
#output - species
```

```python
x = df.iloc[:,:4].values
y = df.iloc[:,4].values
```

```python
#5.Train and Test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state =0)
#Random_state is used to avoid duplicates
```

```python
#6.Normalisation or Scaling
#here step no 6 is not required because our inputs are already scaled
```

```python
#7.Apply a classifier regressor or clusterer
from sklearn.linear_model import LogisticRegression
model = LogisticRegression
```

```python
#8.Fit the model
from sklearn.linear_model import LogisticRegression

# Initialize the model
model = LogisticRegression()

# Fit the model to the training data
model.fit(x_train, y_train)
```

```python
#10.predict the output
y_pred = model.predict(x_test) #using the input testing values , we predict the output.
y_pred
```

```python
y_test #Actual oupit
```

```python
#10.Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
